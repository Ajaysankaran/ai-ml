First starting with a Bigram language model. In Bigram, we only take two characters. We take one letter and check what comes next

```python
for w in words[:3]:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        print(ch1, ch2)
```

In the above code snippet, we append a special Start and End token to denote the start and end of the word

```python
b = {}
# we are storing the occurance of a pair in a dictionary
for w in words:
    chs = ['<S>'] + list(w) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        b[bigram] = b.get(bigram, 0) + 1
```

```python
b.items()
```

b stores the key as a tuple of pair and value as count of number of occurances of the word

```python
sorted(b.items(), key = lambda kv: -kv[1])
```

here we sort the bigram dictionary to see the most occuring bigram

```python
N = torch.zeros((27, 27), dtype=torch.int32)
```

Here 28 * 28 array is used to accomodate all alphabets along with the Start and End delimiter

We need a lookup table to convert characters to Integers

```python
chars = sorted(list(set(''.join(words))))
chars
```

```python
stoi = {s:i+1 for i,s in enumerate(chars)}
stoi['.'] = 0
itos = {i:s for s, i in stoi.items()}
```

```python
# stoi # map of alphabet: number
```

```python
for w in words:
    chs = ['.'] + list(w) + ['.']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        N[ix1, ix2] += 1
```

We get a matrix of occurances of each alphabet pairs

We are going to see the probabilities from the above. From the counts, we can understand how often a word starts with a character

```python
N[0]
```

```python
p = N[0].float()
# convert to float to normalize
```

```python
p = p / p.sum()
p
```

The above is to find the probabilities

```python
p.sum()
```

To sample from these distributions, we are going to use torch.multinomial.
- returns sample from multinomial probability distributions
- We will also use a generator object

```python
g = torch.Generator().manual_seed(2147483647)
ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
itos[ix]
```

Here, using a generator, we are generating random probabilities and using the sanme generator to generate the tensors.
- From the value above, we say that 60% is 0, 30% is 1 and 9% is 2

```python
g = torch.Generator().manual_seed(2147483647)
for i in range(10):
    out = []
    ix = 0
    while True:
        p = N[ix].float()
        p = p / p.sum()
        ix = torch.multinomial(p, num_samples=1, replacement=True, generator=g).item()
        print("ix:", ix)
        out.append(itos[ix])
        if ix == 0:
            break

    print(''.join(out))
```
